#### 1.a. Describe the difference between linear and rank correlation.

Pearson's coefficient is used to measure strength of linear correlation between continuous variables, whereas Spearman's rank correlation is used to measure the strength of correlation between the two ranked variables.

Pearson's coefficient has certain conditions to be met like the data should be normally distributed,
the x and y variables must be continuous, numerical and have a linear relationship.(eg., more time spend on treadmill, more calories we burn)

By using, Spearman's rank correlation, we are not bothered about the x and y variables to be linear and numerical
the only thing we have to keep in mind is, rank correlation applies only when the variables are ordinal.(if a data has values and has an order or rank relationship eg., customer service data: highly satisfied, satisfied, not satisfied)

#### 2.Create a box and whisker plot for the Belize Airbnb data price. In a markdown cell, describe what is going on in as simple of terms as possible.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
belize_df.describe()
```

```python
plt.boxplot(belize_df['price'])
plt.show()
```

In the above plot we can see the outliers, extended upto 12000 representing the source data as is 
and there is no clear picture to see the min, max, median and inner quartile ranges as the boxplot is compressed.  

```python
plt.boxplot(belize_df['price'], sym='', whis=[5,100], showmeans=True, meanline=True, labels=['Price'])
plt.show()
```

The extended line from the box indicates the whiskers, to build the whiskers we measure the inner quartile range i.e 
the distance from 25th percentile to 75th percentile, the lines of the whiskers stretch out until the last data point. 


As the last data point for the price is at 11788, the wishkers extend till the last data point and the box looks still 
compressed.
lets see the more clear boxplot by changing the whis value to 95% to see what is the min price, mean and median. 

```python
plt.boxplot(belize_df['price'], sym='', whis=[0,95], showmeans=True, meanline=True, labels=['Price'])
plt.show()
```

 From the above boxplots, we can expect the minimum price is below 50 dollars and maximum price is nearly between 
11500 to 12000dollars, whereas the median comes around 210 and the average mean falls around 350, approximately.
