#### supervised learning models:
With supervised learning task, we can generate a model or predict an output based on input values.
 In machine learning models the output of the model is typically called a label, target variable, or response variable or dependent variable.
 The inputs into the model are called features or predictors or independent variable.

The algorithm "learns" the optimal model by studying a training set of data that consists of several obsevations, 
or instances, each of which contains values for both features and its labels.here, the supervised learning algorithm acts as
a function that takes training data as input and produces a model as its output.


    

The most popular Python library for machine learning is the Scikit-Learn (sklearn) package. Scikit-Learn provides functions for creating supervised learning models, as well as for performing unsupervised learning tasks,also provides tools for working and preparing data in creating models.
scikit-learn makes it very easy to try different models, since the Train-Test-Split/Instantiate/Fit/Predict pattern applies to all the classifiers and regressors.

Supervised learning models are divided into Classification and Regression problems based on whether the target values are
continuous or categorical.
Classification, when the target values are discrete classes For example, Predicting whether the loan is at 'high risk' or 
'low risk'. In class example like predicting the patient as diabetic or not diabetic.

Regression, when the target values are continuous, real numbers. For example, Predicting a credit score, predicting the sale
price of a home.

#### Data cleaning

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format="%m/%d/%Y %H:%M")
data.head()
```

#### Handling Missing Values

```python
data.isnull().values.any()
```

```python
data.isnull().sum()*100/data.shape[0]
```

we have two columns with missing values('Description', 'CustomerID') 
the 'CustomerID' column, is having almost 25% of the data missing and 'Description' column is hvaing 0.3% of the data missing.

```python
# dropping null features:
data.dropna(inplace=True)
```

```python
data.isnull().sum()*100/data.shape[0]
```

```python
# Firstly i will create a new column as Amount according to the UnitPrice and Quantity purchased.
data['Amount'] = data['Quantity']*data['UnitPrice']
data.head()
```

```python
# Now, will create a new DataFrame with each unique customer and amount they have spent.This is like cluster them according 
# to Amount spent. 

each_customer = data.groupby('CustomerID')['Amount'].sum() #sum()adds up all the amounts and get a total for each CustomerID.
each_customer.reset_index()
```

```python
# Next we can ckeck no. of times customer visited to store by using Invoice column

each_customer_freq = data.groupby('CustomerID')['InvoiceNo'].count()
each_customer_freq.reset_index()
```

```python
# Now we will merge the two DataFrames and store the info under customer_details
customer_details=pd.merge(each_customer_freq,each_customer,on='CustomerID')
customer_details.reset_index()
```

```python
# we can improve this by even getting more details like which customers have visited the store for long time. 
# By this we can cluster the customers who have not visited and can see the underlying pattren. To do this we use
# the column 'InvoiceDate' which contains the datetime with respect to each invoice. we have already converted 
# this column into datetime format.
```

```python
max_date = max(data['InvoiceDate'])
data['recency'] = max_date - data['InvoiceDate']
data.head()
```

```python
# groupby each customer and get minimun from the recency column, last seen customers

dates = data.groupby('CustomerID')['recency'].min()
dates=dates.reset_index()
dates.head()
```

```python
# we use dt.days of the datetime library to get only the days and remove the time
import datetime as dt
dates['recency'] = dates['recency'].dt.days
dates
```

```python
# now, will try to merge this back to earlier 'customer_details' dataset
customer_details=pd.merge(customer_details, dates, on='CustomerID', how='inner')
customer_details
```

```python
# we can see the'Amount' column has many outliers, we need to get rid of them to get good clusters

outlier_detection_x = customer_details.Amount.quantile(0.05)
outlier_detection_y = customer_details.Amount.quantile(0.95)
iqr = outlier_detection_y - outlier_detection_x

customer_details = customer_details[(customer_details['Amount']>=outlier_detection_x-1.5*iqr) & (customer_details['Amount']<=outlier_detection_y+1.5*iqr)]
```

```python
outlier_detection_x = customer_details.InvoiceNo.quantile(0.05)
outlier_detection_y = customer_details.InvoiceNo.quantile(0.95)
iqr = outlier_detection_y - outlier_detection_x

customer_details = customer_details[(customer_details['InvoiceNo']>=outlier_detection_x-1.5*iqr) & (customer_details['InvoiceNo']<=outlier_detection_y+1.5*iqr)]
```

```python
outlier_detection_x = customer_details.recency.quantile(0.05)
outlier_detection_y = customer_details.recency.quantile(0.95)
iqr = outlier_detection_y - outlier_detection_x

customer_details = customer_details[(customer_details['recency']>=outlier_detection_x-1.5*iqr) & (customer_details['recency']<=outlier_detection_y+1.5*iqr)]
```

#### Rescaling the Attributes

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
scaled = sc.fit_transform(customer_details[['InvoiceNo', 'Amount', 'recency']])
scaled.shape
```

```python
scaled_values = pd.DataFrame(scaled)
scaled_values.columns = ['InvoiceNo', 'Amount', 'recency']
scaled_values
```

### K-Means Clustering

```python
# Import KMeans
from sklearn.cluster import KMeans

# # Create a KMeans instance with 3 clusters: model
# model = KMeans(n_clusters=4, max_iter=50)

# # Fit model to points
# model.fit(scaled_values)

# model.labels_

```

```python
scores=[]
for i in range(2,7):
    model = KMeans(n_clusters=i, max_iter=500, verbose=True)
    model.fit(scaled_values)
    scores.append(model.inertia_)
```

```python
len(scores)
```

```python
plt.plot(range(2,7,1), scores)
plt.xticks(ticks=range(2,7))
plt.show()
```

```python
from sklearn.metrics import silhouette_score


for i in range(2,7):
    model = KMeans(n_clusters=i, max_iter=50).fit(scaled_values)
    cluster_labels=model.labels_
    silhouette_avg=silhouette_score(scaled_values,cluster_labels)
    print('For n_clusters{} the silhouette scores is {}'.format(i,silhouette_avg))
```

```python
model = KMeans(n_clusters=3, max_iter=40).fit(scaled_values)
clusters = model.labels_
```

5.	Perform hierarchical clustering for customer segments. What is the ideal number of clusters? Which approach did you use to find the ideal number of clusters and why?



Hierarchical clustering: Process of creating clusters that are predetermined from top to bottom 

Every column begins in a separate cluster, the two closest clusters are merged, at each step and 
continues until all the columns are in a single cluster.This is "agglomerative" hierarchical clustering


```python
# import libraries
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

mergings = linkage(scaled_values, method='complete', metric='euclidean')
dendrogram(mergings,
           labels = clusters,
           leaf_rotation=90,
           leaf_font_size=6)
plt.show()
```

```python
# Single Linkage:
mergings = linkage(scaled_values, method='single', metric='euclidean')
dendrogram(mergings,
           labels = clusters,
           leaf_rotation=90,
           leaf_font_size=6)
plt.show()
```

```python
mergings = linkage(scaled_values, method='average', metric='euclidean')
dendrogram(mergings,
           labels = clusters,
           leaf_rotation=90,
           leaf_font_size=6)
plt.show()
```

```python
#### Cutting the Dendrogram based on K
```

```python
# 3 clusters
from scipy import cluster
cluster_labels =  cluster.hierarchy.cut_tree(mergings, n_clusters=3).reshape(-1, )
cluster_labels
```

```python
customer_details['Cluster_labels'] = cluster_labels
customer_details.head()
```

```python
# Plot Cluster vs Amount

sns.boxplot(x='Cluster_labels', y='Amount', data=customer_details)
```

```python
# Plot Cluster vs Amount

sns.boxplot(x='Cluster_labels', y='InvoiceNo', data=customer_details)
```

```python
# Plot Cluster vs Amount

sns.boxplot(x='Cluster_labels', y='recency', data=customer_details)
```

```python
# Perform the necessary imports
import pandas as pd
from scipy.cluster.hierarchy import fcluster

# Use fcluster to extract labels: labels
labels = fcluster(mergings, 6, criterion='distance')

# Create a DataFrame with labels and varieties as columns: df
df = pd.DataFrame({'labels': labels, 'recency': clusters})

# Create crosstab: ct
ct = pd.crosstab(df['labels'], df['recency'])

# Display ct
print(ct)
```
