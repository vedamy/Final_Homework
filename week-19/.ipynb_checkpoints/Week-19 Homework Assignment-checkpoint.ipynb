### Unsupervised Learning
In unsupervised learning, we work with data sets in which features are provided, but no labels.
Not only are the labels "missing", but we typically do not start an unsupervised learning tasks with a preconception of
what the possible labels might be. The goal of unsupervised learning is to identify structure within the data.

 The algorithm we use in unsupervised learning is K-means clustering.
We start by creating a kmeans model, specifying the number of clusters we want to find, fits the model to the data
(remembers the regions where the different clusters occur) and we use the predict method, which returns a
cluster label for each sample, indicating to which cluster a sample belongs to.

To determine the optimal number of clusters we use inertia, Elbow method and silhouette_score.
When we have tight clusters it is said to e a good cluster, meaning that the samples in each cluster are bunched together, not spread out. To 
measure how the samples are spread out in each cluster we use "inertia". The low values of inertia is preferred.
Its a better to choose an elbow in the inertia plot, the elbow method uses sum of squared distances to choose an 
optimal K-value based on distance between the data points.
Silhouette_score is used to measure the mean of silhoutte coefficient for each sample belonging to different clusters.

#### Data cleaning

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format="%m/%d/%Y %H:%M")
data.head()
```

```python
#### Remove Quantity with negative values
```

```python
data = data[data.Quantity > 0]
data.describe().round(2)
```

So the minimum value for the quantity is 1 i.e, we have managed to remove outlier from the dataset

```python
#### Lets Calculate the Amount Spend by a customer
```

```python
# Firstly i will create a new column as Amount_spent according to the UnitPrice and Quantity purchased.
data['Amount_spent'] = data['Quantity']*data['UnitPrice']
data.head()
```

```python
# Now, will create a new DataFrame with each unique customer and amount they have spent.This is like cluster them according 
# to Amount spent. 

each_customer = data.groupby('CustomerID')['Amount_spent'].sum() #sum()adds up all the amounts and get a total for each CustomerID.
each_customer.reset_index()
```

```python
# Next we can ckeck no. of times customer visited to store by using Invoice column

each_customer_freq = data.groupby('CustomerID')['InvoiceNo'].count()
each_customer_freq.reset_index()
```

```python
# Now we will merge the two DataFrames and store the info under customer_details
customer_details=pd.merge(each_customer_freq,each_customer,on='CustomerID')
customer_details.reset_index()
```

```python
# we can improve this by even getting more details like which customers have visited the store for long time. 
# By this we can cluster the customers who have not visited and can see the underlying pattren. To do this we use
# the column 'InvoiceDate' which contains the datetime with respect to each invoice. we have already converted 
# this column into datetime format.
```

```python
max_date = max(data['InvoiceDate'])
data['recency'] = max_date - data['InvoiceDate']
data.head()
```

```python
# groupby each customer and get minimun from the recency column, last seen customers

dates = data.groupby('CustomerID')['recency'].min()
dates=dates.reset_index()
dates.head()
```

```python
# we use dt.days of the datetime library to get only the days and remove the time
import datetime as dt
dates['recency'] = dates['recency'].dt.days
dates
```

```python
# now, will try to merge this back to earlier 'customer_details' dataset
customer_details=pd.merge(customer_details, dates, on='CustomerID', how='inner')
customer_details
```

```python
# we can see the'Amount' column has many outliers, we need to get rid of them to get good clusters
# the quantile function
outlier_detection_x = customer_details.Amount_spent.quantile(0.05)
outlier_detection_y = customer_details.Amount_spent.quantile(0.95)

# To get the interquartile range for the above quantile function by subtracting 95 percentile with 5 percentile.
iqr = outlier_detection_y - outlier_detection_x

customer_details = customer_details[(customer_details['Amount_spent']>=outlier_detection_x-1.5*iqr) & (customer_details['Amount_spent']<=outlier_detection_y+1.5*iqr)]
```

```python
outlier_detection_x = customer_details.InvoiceNo.quantile(0.05)
outlier_detection_y = customer_details.InvoiceNo.quantile(0.95)
iqr = outlier_detection_y - outlier_detection_x

customer_details = customer_details[(customer_details['InvoiceNo']>=outlier_detection_x-1.5*iqr) & (customer_details['InvoiceNo']<=outlier_detection_y+1.5*iqr)]
```

```python
outlier_detection_x = customer_details.recency.quantile(0.05)
outlier_detection_y = customer_details.recency.quantile(0.95)
iqr = outlier_detection_y - outlier_detection_x

customer_details = customer_details[(customer_details['recency']>=outlier_detection_x-1.5*iqr) & (customer_details['recency']<=outlier_detection_y+1.5*iqr)]
```

#### Rescaling the Attributes

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
scaled = sc.fit_transform(customer_details[['InvoiceNo', 'Amount_spent', 'recency']])
scaled.shape
```

```python
scaled_values = pd.DataFrame(scaled)
scaled_values.columns = ['InvoiceNo', 'Amount_spent', 'recency']
scaled_values
```

4.	Perform k-means on this dataset for customer segments. Customer segments help determine what types of people buy your product, which allows you to target more people like your usual customers. Should you look at all the data, or which subset of data should you use? What is the ideal number of clusters? Which approach did you use to find the ideal number of clusters and why?

We know that Kmeans cluster should have 'k' value for the algorithm ie, the no of clusters. We can find out the optimal value for K with elbow graph.

### K-Means Clustering

```python
# Import KMeans
from sklearn.cluster import KMeans

scores=[]

for i in range(2,7):
    model = KMeans(n_clusters=i, max_iter=500, verbose=True)
    model.fit(scaled_values)
    scores.append(model.inertia_)
```

```python
len(scores)
```

```python
#Elbow graph to find optimal k-value
plt.plot(range(2,7,1), scores)
plt.xticks(ticks=range(2,7))
plt.show()
```

```python
#Silhouette_score to find optimal k-value
from sklearn.metrics import silhouette_score


for i in range(2,7):
    model = KMeans(n_clusters=i, max_iter=50).fit(scaled_values)
    cluster_labels=model.labels_
    silhouette_avg=silhouette_score(scaled_values,cluster_labels)
    print('For n_clusters{} the silhouette scores is {}'.format(i,silhouette_avg))
```

From the above elbow graphs and silhouette scores we can take ‘3’ as our k value for number of clusters.

```python
# Final model with 3 clusters
model = KMeans(n_clusters=3, max_iter=40).fit(scaled_values)
clusters = model.labels_
```

```python
customer_details['Cluster']=clusters
customer_details
```

With the help of boxplots we can visualize the clusters formed on different features

```python
# Clusters based on Amount Spent

sns.boxplot(x='Cluster', y='Amount_spent', data=customer_details)
plt.show()
```

```python
# Clusters based on customer last seen

sns.boxplot(x='Cluster', y='InvoiceNo', data=customer_details)
plt.show()
```

```python
#Clusters based on No of Transactions

sns.boxplot(x='Cluster', y='recency', data=customer_details)
plt.show()
```

5.	Perform hierarchical clustering for customer segments. What is the ideal number of clusters? Which approach did you use to find the ideal number of clusters and why?



Hierarchical clustering: Every column begins in a separate cluster, the two closest clusters are merged, at each step and continues until all the columns are in a single cluster.This is "agglomerative" hierarchical clustering 

```python
# import libraries
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

mergings = linkage(scaled_values, method='complete', metric='euclidean')
dendrogram(mergings,
           labels = clusters,
           leaf_rotation=90,
           leaf_font_size=6)
plt.show()
```

```python
mergings = linkage(scaled_values, method='average', metric='euclidean')
dendrogram(mergings,
           labels = clusters,
           leaf_rotation=90,
           leaf_font_size=6)
plt.show()
```

```python
# Cutting the Dendrogram based on K
from scipy import cluster
cluster_labels =  cluster.hierarchy.cut_tree(mergings, n_clusters=3).reshape(-1, )
cluster_labels
```

```python
# Assigning cluster labels
customer_details['Cluster_labels'] = cluster_labels
customer_details.head()
```

With the help of boxplots we can visualize the clusters formed on different features:

```python
# Plot Cluster vs Amount_spent

sns.boxplot(x='Cluster_labels', y='Amount_spent', data=customer_details)
```

```python
# Plot Cluster vs InvoiceNo

sns.boxplot(x='Cluster_labels', y='InvoiceNo', data=customer_details)
```

```python
# Plot Cluster vs recency

sns.boxplot(x='Cluster_labels', y='recency', data=customer_details)
```
