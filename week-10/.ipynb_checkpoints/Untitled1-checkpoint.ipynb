1.Write a function that calculates the dot product manually AND a function that calculates the cross product manually. Both should accept two vectors and perform that calculation. Be sure to use try except and error handling for basic issues that might occur.

```python
import pandas as pd
import numpy as np
```

```python
def dot_product (a,b):
    y = 0
    for x in range(a.shape[0]):
        try:
            y += a[x]* b[x]
        except Exception as e:  # try except Exception as e for error handling issues
            print(e)
    else: 
        print(y)
```

```python
v1= np.array([1, 0, 2, 7])
v2 = np.array([3, 8, 5, 3])

dot_product(v1, v2)
```

```python
np.dot(v1, v2)
```

#### what is one hot encoding and why we use it: 

It is a process used to convert data(human readable format) into machine readable format which is generally numeric and binary.
The best way to do this is by one hot encoding and we call the process as converting categorical variables into binary variables.

Performing one hot encoding requires data to be numeric, if the given data is categorical then we must encode it to numbers before applying one hot encoding, which is subsequently used to fit and evaluate a model and makes more expressive.
