6.	Create subplot histograms to show if loan_amounts differ by region. Summarize your findings.

```python
loan_amt = df[['state', 'loan_amount']].dropna()
```

```python
regions = {
'AK' : 'West',
'AL' : 'South',
'AR' : 'South',
'AZ' : 'West',
'CA' : 'West',
'CO' : 'West',
'CT' : 'North East',
'DC' : 'West',
'DE' : 'South',
'FL' : 'South',
'GA' : 'South',
'HI' : 'West',
'ID' : 'West',
'IL' : 'Midwest',
'IN' : 'Midwest',
'KS' : 'Midwest',
'KY' : 'South',
'LA' : 'South',
'MA' : 'North East',
'MD' : 'South',
'ME' : 'North East',
'MI' : 'Midwest',
'MN' : 'Midwest',
'MO' : 'Midwest',
'MS' : 'South',
'MT' : 'Midwest',
'NC' : 'South',
'ND' : 'Midwest',
'NE' : 'Midwest',
'NH' : 'North East',
'NJ' : 'North East',
'NM' : 'West',
'NV' : 'West',
'NY' : 'North East',
'OH' : 'Midwest',
'OK' : 'South',
'OR' : 'West',
'PA' : 'North East',
'RI' : 'North East',
'SC' : 'South',
'SD' : 'Midwest',
'TN' : 'South',
'TX' : 'South',
'UT' : 'West',
'VA' : 'South',
'VT' : 'North East',
'WA' : 'West',
'WI' : 'Midwest',
'WV' : 'South',
'WY' : 'West'
}

def get_region(state):
    return regions[state]

loan_amt['Region'] = loan_amt['state'].apply(get_region)
loan_amt
```

```python
state_NE = loan_amt.loc[loan_amt.Region=='North East', 'loan_amount']
state_W = loan_amt.loc[loan_amt.Region=='West', 'loan_amount']
state_MW = loan_amt.loc[loan_amt.Region=='Midwest', 'loan_amount']
state_S = loan_amt.loc[loan_amt.Region=='South', 'loan_amount']

kwargs = dict(alpha=0.5, bins=50)

plt.hist(state_NE, **kwargs, color ='g', label='North East Region')
plt.hist(state_W, **kwargs, color = 'b', label='West Region')
plt.hist(state_MW, **kwargs, color = 'r', label='MidWest Region')
plt.hist(state_S, **kwargs, color = 'c', label='South Region')
plt.gca().set(title='loan_amounts differ by region', ylabel='Frequency')
plt.xlim(0, 50000)
plt.legend()
plt.show()
```

```python
state_NE = loan_amt.loc[loan_amt.Region=='North East', 'state']
state_W = loan_amt.loc[loan_amt.Region=='West', 'state']
state_MW = loan_amt.loc[loan_amt.Region=='Midwest', 'state']
state_S = loan_amt.loc[loan_amt.Region=='South', 'state']

kwargs = dict(alpha=0.5, bins=100)

plt.hist(state_NE, **kwargs, color ='g', label='North East Region')
plt.hist(state_W, **kwargs, color = 'b', label='West Region')
plt.hist(state_MW, **kwargs, color = 'r', label='MidWest Region')
plt.hist(state_S, **kwargs, color = 'c', label='South Region')
plt.gca().set(title='loan_amounts differ by region', ylabel='Frequency')
plt.xlim(0,55)
plt.legend()
plt.show()
```

```python
# Plot
fig, axes = plt.subplots(1, 4, figsize=(10,2.5), dpi=100, sharex=True, sharey=True)
colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:pink']

for i, (ax, Region) in enumerate(zip(axes.flatten(), loan_amt.loan_amount.unique())):
    state_NE = loan_amt.loc[loan_amt.Region=='North East', 'loan_amount']
    state_W = loan_amt.loc[loan_amt.Region=='West', 'loan_amount']
    state_MW = loan_amt.loc[loan_amt.Region=='Midwest', 'loan_amount']
    state_S = loan_amt.loc[loan_amt.Region=='South', 'loan_amount']

    ax.hist(state_NE, state_W, state_MW, state_S, alpha=0.5, bins=100, density=True, stacked=True, label=str(Region), color=colors[i])
    ax.set_title(Region)

plt.suptitle('loan_amounts differ by region', y=1.05, size=16)
ax.set_xlim(50, 70); ax.set_ylim(0, 1);
plt.tight_layout()
```
