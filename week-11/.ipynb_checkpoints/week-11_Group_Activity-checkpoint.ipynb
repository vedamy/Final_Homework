```python
x = df['Coronavirus: (United States)'].values
x.size
```

```python
#80% of my data for training and 20% of the data is for testing
train = x[0:42] # 30 data points(weeks) as train data
test = x[42:]  #20 dataponts (weeks) as test data
predictions = []
a=train.size
b=test.size
print(a)
print(b)
```

### Autoreggresive AR Model

```python
#from statsmodels.tsa.ar_model import AR
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error

# train the forcasting model
model_ar = AutoReg(train,lags=1)
model_ar_fit = model_ar.fit() 
```

```python
predictions = model_ar_fit.predict(start=42, end=52)
predictions
```

```python
plt.plot(test)
plt.plot(predictions, color = 'red')
```

From the below plot we can observe that the predictions do not line up ith the actual test values which indicate that
the accuracy of the model is not good. This can be attributed to the non-stationary nature of the data that was sourced.

Poor model prediction may also be contributed to the vaccination factor not being considered in this case.

## ARIMA 

ARIMA Model : Combination of three techniques which is auto regressive, integrated and moving average

```python
#from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA
# p,d,q 
# p = periods taken for autoregressive model,
# d- integrated order, difference
# q - periods in moving average model
model_arima = ARIMA(train, order=(3,1,1))
model_arima_fit = model_arima.fit()
print(model_arima_fit.aic)
```

```python
predictions = model_arima_fit.forecast(steps=10)
predictions
```

### ARIMA Plot:


```python
plt.plot(test)
plt.plot(predictions, color = 'red')
```


 From the below ARIMA model we can infer that compared to the AR model, ARIMA model's prediction accuracy is relatively higher.
    
 The AIC value indications the accuracy of the forecast.

 We can play with P,D,Q values in the ARIMA model to fine tune the accuracy

#### 2.What is a fourier transform? Perform one on y = np.cos(x/2), and both show (with code) and explain the key elements.

 Fourier transform is a technique which decomposes a signal into its constituent frequencies.

 The first plot below time v/s intensity represents the signal in time domain where time is plotted on x-axis and amplitude 
 on y-axis.

```python
from scipy import signal

x = np.arange(1, 100, 1)
y = np.cos(x/2)
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Intensity')
plt.show()
```

 The below plot represents a signal in frequency domain where frequency is plotted on x-axis and Intensity or power is 
plotted on Y-axis

```python
#power spectral density
psd = np.abs(np.fft.rfft(y))
freqs = np.fft.rfftfreq(len(y))
plt.plot(freqs,psd)
#amplitude
plt.axvline(1/(6*np.pi),c='y')
plt.xlabel('Frequency')
plt.ylabel('Intensity')
plt.show()
```
