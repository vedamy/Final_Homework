```python
temp = pd.read_csv(r"A:\launch_code_STL\Final_Homework\week-11\temperatures\temperature.csv", parse_dates=['datetime'])

temp['month'] = pd.DatetimeIndex(temp['datetime']).month
season = []
for value in temp["month"]:
    if value in [6,7,8,9]:
        season.append("Summer")
    elif value in [11,12,1,2,3,4]:
        season.append("Winter")
    else:
        season.append("Fall or Spring")
temp["Season"] = season 

season_avg = temp.groupby('Season').mean().reset_index()
# season_avg
df_season= season_avg[season_avg['Season'].isin(['Summer', 'Winter'])]
dft=df_season.T.iloc[1:37].reset_index()
dft.rename(columns = {'index':'City', 1 : 'Avg_Summer_temp', 2 : 'Avg_Winter_temp'}, inplace = True)
dft
#Below is the List of cities with Average summer and winter temperatures:
```

```python
import pandas as pd
Summmer_check = []
Winter_check = []
#creating a column for checking summer condition
for i in range (0,35):
    for j in range (0,35):
        if dft.iloc[i, 1] > dft.iloc[j, 1]:
            result_summer=dft.iloc[i,0],dft.iloc[j,0]
            Summmer_check.append(result_summer)
        else:
            Summmer_check.append("summer_False")
#creating a column for checking winter condition
for i in range (0,35):
    for j in range (0,35):
        if dft.iloc[i, 2] < dft.iloc[j, 2]:
            result_winter=dft.iloc[i,0],dft.iloc[j,0]
            Winter_check.append(result_winter)
        else:
            Winter_check.append("Winter_False")

# Merging two colum conditions into a single dataframe            
Final=pd.DataFrame(list(zip(Summmer_check,Winter_check)),columns =['Summer_chk','Winter_chk'])
Result=Final.loc[Final['Summer_chk'] == Final['Winter_chk']]

Result
```

The above cell results in cities where former city is warmer than later city in the summer, but also colder 
the winter, on average.

6.	For two of the cities you found in question 5 (there are many possible answers), clean up the data via rolling averages and plot a clear visual. Explain how rolling average differs from moving average (MA) model

```python
temp = pd.read_csv(r"A:\launch_code_STL\Final_Homework\week-11\temperatures\temperature.csv", parse_dates=['datetime'])
temp
```

```python
Daily_temp=temp.resample('M', on = 'datetime').mean()
Daily_temp
```

```python
Cities = pd.DataFrame(Daily_temp, columns = ['Portland', 'San Francisco']).reset_index()
Cities.rename(columns = {'datetime':'Date','Portland':'CityA', 'San Francisco':'CityB'}, inplace = True)
Cities
```

```python
# Daily_temp['MA_Cities'] = Daily_temp.Portland.rolling(21).mean()
Cities['QMA_Portland'] = Cities.CityA.rolling(3).mean()
Cities['QMA_San_Francisco'] = Cities.CityB.rolling(3).mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.grid(True)
# plt.plot(Daily_temp['MA_Cities'], label='Portland')
plt.plot(Cities['QMA_Portland'], label='CityA(Portland):Quarterly Moving Average')
plt.plot(Cities['QMA_San_Francisco'], label='CityB(San Francisco):Quarterly Moving Average')
plt.legend(loc=2)
```

The Rolling and Moving Averages are used to analyze the trends in the data for specific time series.

"Rolling average" includes the data which is up-to-date by continuously updataing the data.

For example, the rolling average of Cities temperature at October 2012 would be calculated by adding the temperatures in August,
September, and october and then divide that sum by three.

Whereas the average of a dataset for over a specific period of time gives the "Moving average".

For example, the Moving average of Cities temperature at October 2012 with specified period of two would be caliculated
by adding the temperatures in September and october and then divide that sum by two.


7.	Perform SARIMA on five cities, iterating through the cities programmatically. Explain your results.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
from pandas.plotting import register_matplotlib_converters
from statsmodels.tsa.arima_model import ARMA
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
register_matplotlib_converters()
from time import time
```

```python
def sarima_city_temp(select_city): 
#     for i in range (1, i):
#         if i <= 5:
            City=Daily_temp[select_city]
            City_diff = City.diff(periods=1)
            print(City.head())

            # Removing NaN values
            City_diff_new = City_diff[1:]
            City_diff_new.tail()

            plt.plot(City_diff_new)

            from statsmodels.graphics.tsaplots import plot_acf
            plot_acf(City_diff_new, lags = 15, missing='drop')

            train_end = datetime(2017,10,31)
            test_start = datetime(2017,11,1)
            #specify train and test data
            train = City[:train_end]
            test = City[test_start:]

            my_temp = (0,1,1)  # p,d,and q
            my_seasonal_temp = (2, 1, 0, 12) # p,d,q,and s
            # define model
            model = SARIMAX(train, order=my_temp, seasonal_order=my_seasonal_temp)

            #fit the model
            start = time()
            model_fit = model.fit()
            end = time()
            print('Model Fitting Time:', end - start)

            model_fit = model.fit()
            print(model_fit.summary())
            # Make Prediction 
            predictions = model_fit.forecast(len(test))
            predictions = pd.Series(predictions,index=test.index)

            # plot graph
            plt.figure(figsize=(18, 5))
            plt.plot(predictions, color='red', label='Predicted Temp')
            plt.plot(test, color='black', label='Actual Temp')
            plt.title('Actual Vs Forescated Temperature for 30 Day period Using SARIMA Model', fontsize=20)
            plt.ylabel('Temperature', fontsize=20)
            plt.legend()
            plt.show()

    #         print(City.head(5))
#         else:
#             print('Ensure to provide Numbers only between 1 and 5')
            return


sarima_city_temp('Portland')
```
