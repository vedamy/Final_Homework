```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
from pprint import pprint
```

```python
url = 'https://www.mercadolibre.com.mx/'
page=requests.get(url)
page
```

```python
soup = bs(page.text,'html.parser')
# print(soup.prettify())
```

```python
print(soup.title)
```

```python
for i in soup.find_all('a', href = True):
    print(i['href'])
    if "www.mercadolibre.com.mx/internetmoneyrecords/flossin" in i['href']:
        nextpage = request.grt(i['href'])
        nextsoup = bs(nextpage.text, 'html.parser')
        
        print("next url title: ", nextsoup.find('title').string)
```

```python
url = 'https://www.mercadolibre.com.mx/ofertas'
page=requests.get(url)
soup = bs(page.text,'html.parser')

div_prod = soup.find_all('div', class_='promotion-item__container')
# div_prod
```

```python
Price_tracker=[]
for div in div_prod:
    price = div.find('span', class_='promotion-item__price').text
    item_title = div.find('p', class_='promotion-item__title')
    product_name = item_title.text
    product_price = {'item':product_name,'Price':price}
    Price_tracker.append(product_price)
        
df = pd.DataFrame(Price_tracker)
df       
```
