```python
nulls = df.isnull().sum().sort_values(ascending = False)
prcet = round(nulls/len(df)*100,2)

df_null = pd.DataFrame(columns =  ['Attr','Total','Percent'])
df_null.Attr  = nulls.index
df_null.Total = nulls.values
df_null.Percent = prcet.values
# print(df_null.tail(50))
```

```python
# Drop attributes with more than 30% data missing

for i in df_null.Attr[df_null['Percent'] > 10]:
    df = df.drop([i], axis=1)
#     print(df.shape,i)
```

```python
#Fill remaining NULLs with mode values
for i in df.columns:
    if df[i].isnull().sum() > 0:
        df[i].fillna(df[i].mode().mean(), inplace=True)
#         print('filled',i)
```

```python
#Re-check NULL Percentages
nulls = df.isnull().sum().sort_values(ascending = False)
prcet = round(nulls/len(df)*100,2)

df_null = pd.DataFrame(columns =  ['Attr','Total','Percent'])
df_null.Attr  = nulls.index
df_null.Total = nulls.values
df_null.Percent = prcet.values
print(df_null.head())
```

3.	Perform linear regression on the dataset to see if you can predict water temperature based on salinity
