```python
# 1.Make a function that can take any non-negative integer as an argument and return it with its digits in descending order.
# Essentially, rearrange the digits to create the highest possible number. If no number is passed in, make the default value
# inputted = 111111.
# Input = 43905 Output = 95430
# Input = 584366 Output = 866543
```

```python
def des_order(new_string = 11111):
    new_string = str(new_string)
    list_A = []
    list_B = ""
    for value in new_string:
        list_A.append(value)
        list_A.sort(reverse = True)
    print(list_A)
    for value in list_A:
        list_B += value
    return list_B
```

```python
# How many unique shows were there each year?
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_to_load = "C:/Users/rsagu/Desktop/launch_code/Data-Science-Assignments/course_material/week_05/JEOPARDY_CSV.csv"

jeopardy = pd.read_csv(file_to_load)

jeopardy.rename(columns={
    ' Air Date': 'Date',
    ' Round': 'Round',
    ' Category': 'Category',
    ' Value': 'Value',
    ' Question': 'Question',
    ' Answer': 'Answer'
}, inplace=True)
print(jeopardy.head())
```

```python
jeo = jeopardy[['Show Number', 'Date' ]]
jeo.head()
```

```python
#Remove duplicates on shownumber and air date to get distinct values
x = jeo.drop_duplicates(subset=['Show Number', 'Date'], keep='last')
x
```

```python
# Sorting by columns 
x.sort_values(by=['Show Number'])
```

```python
x.groupby('Show Number').sum()
```

```python
from datetime import datetime
x['year'] = pd.DatetimeIndex(x['Date']).year
x
```

```python
stat = x.groupby('year').count()

stat['year'] = stat.index
stat
```
